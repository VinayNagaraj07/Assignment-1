## Q11)  Suppose we want to estimate the average weight of an adult male in    Mexico. We draw a random sample of 2,000 men from a population of 3,000,000 men and weigh them. We find that the average person in our sample weighs 200 pounds, and the standard deviation of the sample is 30 pounds. Calculate 94%,98%,96% confidence interval?

- n=2000
- Sample Mean=200
- Sample S.D=30
- N = 3,000,000

```python
n=2000 # sample size
sample_mean=200
sample_SD=30
```

- 94% confidence interval

```python
confidence_interval = .94
area=(1+confidence_interval)/2
area
```

- for C.I 94% , for 0.97 we get Z-value = 1.88(Appx)

```python
z_value=1.88
```

## Margin of error=Z-value*(S.D/Sqrt(n))

```python
MoE=z_value*(sample_SD/np.sqrt(n))
MoE
```

```python
range_average=(sample_mean-MoE,sample_mean+MoE)
range_average
```

- 98% confidence interval

```python
confidence_interval = .98
area=(1+confidence_interval)/2
area
```

- for C.I 98% , for 0.99 we get Z-value = 2.33(Appx)

```python
z_value=2.33
```

```python
MoE=z_value*(sample_SD/np.sqrt(n))
MoE
```

```python
range_average=(sample_mean-MoE,sample_mean+MoE)
range_average
```

- 96% confidence interval

```python
confidence_interval = .96
area=(1+confidence_interval)/2
area
```

- for C.I 96% , for 0.98 we get Z-value = 2.06(Appx)

```python
z_value=2.06
```

## Q 24)   A Government  company claims that an average light bulb lasts 270 days. A researcher randomly selects 18 bulbs for testing. The sampled bulbs last an average of 260 days, with a standard deviation of 90 days. If the CEO's claim were true, what is the probability that 18 randomly selected bulbs would have an average life of no more than 260 days

- Null Hypothesis (H0) - sample_Mean>270
- Alternate Hypothesis (Ha) - Pop_Mean(Mu)<=270

```python
pop_mean=270
n=18
samp_mean=260
samp_std=90
```

```python
t_value=(samp_mean-pop_mean)/(samp_std/np.sqrt(18))
```

```python
t_value
```

```python
p_lesser_bulb=sc.stats.t.cdf(abs(t_value),n-1)#P(x<260)
```

```python
p_value_bulb=1-p_lesser_bulb
p_value_bulb
```

```python
if p_value_bulb < 0.05:
    print("Null Hypothesis is rejected , Hence the assumption made is Correct")
else:
    print("Null Hypothesis has been failed to reject , Hence the assumption made is not correct")
```
